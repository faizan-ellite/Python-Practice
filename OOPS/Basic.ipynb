```python
class BankAccount:

    def __init__(self, customer_name, balance):
        self.customer_name = customer_name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"₹{amount} deposited.")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"₹{amount} withdrawn.")
        else:
            print("Insufficient Balance")

    def check_balance(self):
        print(f"Current Balance: ₹{self.balance}")


```

```python
account = BankAccount("sumit", 10000)

account.deposit(5000)
account.withdraw(15000)
account.check_balance()
```

## What is Abstraction in Python?

Abstraction means hiding complex implementation details and showing only the necessary features to the user.

Real-life Example

When you use an ATM:

You enter your card and PIN.
You click "Withdraw Money".
You get cash.

You don't see how the bank server verifies your account, checks balance, updates databases, and records transactions. Those details are hidden.

That's Abstraction.

## Encapsulation in Python: Restricting direct access to attributes & methods
Encapsulation restricts access to certain attributes or methods to protect the data and
enforce controlled access.

```python
class Student:
    def __init__(self, name, grade, percentage):
        self.name = name
        self.grade = grade
        self.__percentage = percentage # Private attribute (hidden)
        
    def get_percentage(self): # Public method to access the private attribute
        return self.__percentage
```

```python
# Creating a student object
Elite = Student("Sumit", 10, 98)
# Accessing the private attribute using the public method
print(f"{Elite.name}'s percentage is {Elite.get_percentage()}%.")
```

## Inheritance in Python: Reusing Parent’s prop & methods
Inheritance (parent-child), allows one class (child) to reuse the properties and methods of
another class (parent). This avoids duplication and helps in code reuse.

```python
class Student:
    def __init__(self, name, grade, percentage):
        self.name = name
        self.grade = grade
        self.percentage = percentage
    def student_details(self): # method
       print(f'{self.name} is in {self.grade} grade with {self.percentage}%')
```

```python
class GraduateStudent(Student): # GraduateStudent inherits from Student

    def __init__(self, name, grade, percentage, stream):
        super().__init__(name, grade, percentage) # Call parent class initializer
        self.stream = stream # New attribute specific to GraduateStudent

    def student_details(self):
        super().student_details()
        print(f"Stream: {self.stream}")
```

```python
Elite = Student("Sumit",10,98)
Elite.student_details()
```

```python
Elite = GraduateStudent("Sumit",10,98,"Banking")
Elite.student_details()
```
